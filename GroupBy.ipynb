```python
import pandas as pd

stats = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/laliga.csv')
stats
```

```python
red_cards_by_team = (
    stats.groupby('Team')['Red Cards']
    .sum()
    .nlargest(5)
)

red_cards_by_team
```

```python
avg_long_passes = (
    stats.groupby('Position')['Long passes']
    .mean()
)

avg_long_passes
```

```python
top_shirt_goals = (
    stats.groupby('Shirt number')['Goals scored']
    .sum()
    .nlargest(10)
)

top_shirt_goals
```

```python
shots_df = (
    stats.groupby('Team')
    .agg(
        total=('Shots', 'sum'),
        on_target=('Shots on target', 'sum')
    )
)

shots_df
```
