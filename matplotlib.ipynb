```python
plants=['spinach','turnip','rhubarb','broccoli','kale']
died=[10,25,5,30,45]
germinated=[45,56,78,46,97]
plt.bar(plants,germinated,color='green')
plt.title('germination rate')
```

```python
plt.figure(figsize=(6,6) )
plt.bar(plants,died,color='red')
plt.title('failure rates')
plt.ylabel('number of dead seedlings')
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants,died)
plt.bar(plants,germinated)
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants,died,color='#e74c3c',bottom=germinated)
plt.bar(plants,germinated,color='#2ecc71')
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants,died,color='#e74c3c',bottom=germinated,label='failed seedlings')
plt.bar(plants,germinated,color='#2ecc71',label='germinated seedlings')
plt.legend(shadow=True,frameon=True,facecolor='white')
plt.title('Seedlings')
```
