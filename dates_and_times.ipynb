```python
houses = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/kc_house_data.csv',parse_dates=['date'])
houses
```

```python
days_span = (houses["date"].max() - houses["date"].min()).days
days_span
```

```python
most_sales_date = houses["date"].value_counts().idxmax()

houses[houses["date"] == most_sales_date]
```

```python
houses["year"] = houses["date"].dt.year
houses["year"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    title="Home Sales by Year"
)
```

```python
houses_2014 = houses[houses["date"].dt.year == 2014]

houses_2014.sort_values("price", ascending=False)
```

```python
houses["month"] = houses["date"].dt.month

monthly_sales = houses["month"].value_counts().sort_index()
monthly_sales
```
